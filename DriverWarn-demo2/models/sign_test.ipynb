```python
from __future__ import absolute_import
from __future__ import division
from __future__ import print_function

import numpy as np

from PIL import Image,ImageDraw

import tensorflow as tf # TF2


if __name__ == "__main__":
    sign_model_path = "./traffic_sign_detect.tflite"
    img_path = "./img/00670.png"

    sign_interpreter = tf.lite.Interpreter(model_path=sign_model_path)
    sign_interpreter.allocate_tensors()

    input_details = sign_interpreter.get_input_details()
    print("input details:",input_details)
    output_details = sign_interpreter.get_output_details()

    print("output details:",output_details)

    sign_float = input_details[0]['dtype'] == np.float32

    sign_height = input_details[0]['shape'][1]
    sign_width = input_details[0]['shape'][2]

    sign_img = Image.open(img_path).resize((sign_width,sign_height))
    sign_img.show()
    sign_draw = ImageDraw.Draw(sign_img)

    sign_data = np.expand_dims(sign_img, axis=0)

    if sign_float:
        sign_data = (np.float32(sign_data) - 128) / 256 

    sign_interpreter.set_tensor(input_details[0]['index'], sign_data)

    sign_interpreter.invoke()

    output_data = sign_interpreter.get_tensor(output_details[0]['index'])
    results = np.squeeze(output_data)
    print("results : ",results,"\n======================================")

    output_data1 = sign_interpreter.get_tensor(output_details[1]['index'])
    results1 = np.squeeze(output_data1)
    print("results---1--- : ",results1,"\n======================================")

    output_data2 = sign_interpreter.get_tensor(output_details[2]['index'])
    results2 = np.squeeze(output_data2)
    print("results---2--- : ",results2,"\n======================================")

    output_data3 = sign_interpreter.get_tensor(output_details[3]['index'])
    results3 = np.squeeze(output_data3)
    print("results---3--- : ",results3,"\n======================================")

    for point in range(3):
        top_left_x = int(output_data[0][point][1] * sign_width)
        top_left_y = int(output_data[0][point][0] * sign_height)
        bottom_right_x = int(output_data[0][point][3] * sign_width)
        bottom_right_y = int(output_data[0][point][2] * sign_height)

        print(top_left_x,top_left_y,bottom_right_x,bottom_right_y)

        sign_draw.rectangle([top_left_x,top_left_y,bottom_right_x,bottom_right_y],outline="green")


    sign_img.show()








```
